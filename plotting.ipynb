```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = sns.load_dataset("tips")
df.head()
```

We import Seaborn (sns) and Matplotlib (plt).

Load the tips dataset, which contains restaurant bill and tip details.

.head() shows the first 5 rows → useful to understand the structure of the dataset.

```python
sns.countplot(x="day", data=df)
plt.title("Customer Count by Day")
plt.show()
```

A countplot is used for categorical data.

This shows how many customers came on Thur, Fri, Sat, Sun.

Observation: Weekends (Sat & Sun) have more customers, Thursday has the least.

```python
sns.histplot(df["total_bill"], bins=20, kde=True, color="skyblue")
plt.title("Distribution of Total Bill")
plt.xlabel("Total Bill Amount")
plt.show()
```

A histogram is used for continuous numeric data.

Most total bills are between 10 to 20 dollars.

The KDE curve shows a smooth distribution.

Some customers spent over $40, which are outliers.

```python
sns.boxplot(x="day", y="total_bill", data=df)
plt.title("Spread of Total Bill by Day")
plt.show()
```

A boxplot shows distribution + outliers.

The median line (inside the box) shows typical bill size.

Weekends (Sat, Sun) have a larger spread in total bills.

Outlier points (dots above whiskers) show very high bills.

```python
sns.scatterplot(x="total_bill", y="tip", hue="time", style="sex", data=df)
plt.title("Relationship between Total Bill and Tip")
plt.show()
```

A scatterplot is used to check relationships between two continuous variables.

Higher bills → higher tips (positive correlation).

Dinner customers tend to tip more than lunch customers.

Male & female customers show slightly different tipping patterns.
